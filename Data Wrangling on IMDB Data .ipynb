# lets do some join

```python
#lsts us create a dataframe which consists of onlyenglish movies
```

```python
eng_movies=df[df['original_language']=='en']
eng_movies
```

```python
long_movies=df.loc[df['runtime']>=100,['imdb_id','original_language']]
long_movies
```

```python
eng_movies.shape,long_movies.shape
```

```python
type(long_movies)
```

# Using inner join to merge the two dataframes

```python
merge_df= pd.merge(eng_movies,long_movies,how='inner',on='imdb_id')
merge_df
```

```python
merge_df.shape
```

```python
merge_df= pd.merge(eng_movies,long_movies,how='left',on='imdb_id')
merge_df
```

```python
merge_df= pd.merge(eng_movies,long_movies,how='right',on='imdb_id')
merge_df
```

```python
merge_df= pd.merge(eng_movies[['imdb_id','runtime']],long_movies,on='imdb_id',how= 'outer')
merge_df
```
